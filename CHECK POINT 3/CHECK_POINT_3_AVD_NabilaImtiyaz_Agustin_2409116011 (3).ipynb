```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/Praktikum AVD/top_expensive_leagues.csv")

df
```

```python
#analisis liga dengan basis penggemar terbesar
league_viewers = df.groupby("League Name")["Viewership"].sum().reset_index()
league_viewers = league_viewers.sort_values(by="Viewership", ascending=False).head(10)

plt.figure(figsize=(12, 6))
ax1 = sns.barplot(x="Viewership", y="League Name", data=league_viewers, color="skyblue")

for index, value in enumerate(league_viewers["Viewership"]):
    ax1.text(value + 0.01 * max(league_viewers["Viewership"]), index, f"{value:,}",
             va='center', fontsize=10)

plt.xlabel("Total Viewers")
plt.ylabel("League")
plt.title("Top 10 Leagues with the Largest Viewership")
plt.show()
```

```python
#analisis 10 negara dengan viewership terbesar
top_countries = df.groupby("Country")["Viewership"].sum().reset_index()
top_countries = top_countries.sort_values(by="Viewership", ascending=False).head(10)
plt.figure(figsize=(12, 6))
ax2 = sns.barplot(x="Viewership", y="Country", data=top_countries, color="pink")

for index, value in enumerate(top_countries["Viewership"]):
    ax2.text(value + 0.01 * max(top_countries["Viewership"]), index, f"{value:,}", va='center', fontsize=10)

plt.xlabel("Total Viewers")
plt.ylabel("Country")
plt.title("Top 10 Countries with the Largest Viewership")
plt.show()
```
