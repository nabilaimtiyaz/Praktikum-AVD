4. Konstruksi Data (Construct Data) (penambahan kolom)

```python
#(tahun saat ini)
current_year = 2025

#(tambah kolom Revenue per Team(USD) )
df['Revenue per Team (USD)'] = df['Revenue (USD)'] / df['Total Teams']

#(tambah kolom League Age)
df['League Age'] = current_year - df['Founded Year']

#(tambah kategori pendapatan berdasarkan kuartil)
q1 = df['Revenue (USD)'].quantile(0.33)
q2 = df['Revenue (USD)'].quantile(0.66)

def categorize_revenue(value):
    if value <= q1:
        return "Low"
    elif value <= q2:
        return "Medium"
    else:
        return "High"

df['Revenue Category'] = df['Revenue (USD)'].apply(categorize_revenue)

print(df.head())
```
